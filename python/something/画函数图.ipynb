```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(0.01,1,0.001)
# y = [math.log(a) for a in x]
y_1 = [-math.log(a) for a in x]
y_2 = [math.log(1-a) for a in x]
plt.title("y=log(x)")
# plt.plot(x,y)
plt.plot(x,y_1)
plt.plot(x,y_2)
plt.show()
%matplotlib inline
```

```python
x = np.arange(0.01,1,0.001)
y = [math.sin(3*a)+13*math.cos(13*a) for a in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0.01,100,0.001)
y = [math.e**(-t) for t in x]
plt.plot(x,y)
plt.show()
```
