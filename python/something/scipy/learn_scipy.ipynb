```python
import scipy.stats as stats
x =stats.uniform.rvs(size=20) #产生20个在[0,1]均匀分布的随机数
print(x) 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(x)
plt.show()
```

```python
y = stats.norm.rvs(size=20,loc=0,scale=1)
z = stats.poisson.rvs(0.6,loc=0,size=20)
print(y)
print(z)
plt.plot(y)
plt.plot(z)
plt.show()
```
